```python
import pandas as pd
import numpy as np
url="https://raw.githubusercontent.com/wangjk321/Intragenic-cohesin-project/master/March2020/bino161_normalizedContinuous15.matrix"
mt = pd.read_csv(url,sep="\t")
mt.index=mt.CohesinPos
mt.drop("CohesinPos",axis=1,inplace=True)
```

```python
intraMT= mt.loc[mt["cohesin.intra"]==1]
```

```python
X = intraMT.drop(["Mvalue","Mvalue_continue","cohesin.TSS","cohesin.TES","cohesin.extraFar","cohesin.extraNear","cohesin.intra"],
                 axis=1)
y= intraMT["Mvalue_continue"]
```

```python
X.shape,y.shape
```

```python
## 1. Recursive feature elimination with cross-validation
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFECV
lr = LinearRegression()
rfecv = RFECV(estimator=lr, step=1, cv=4,scoring='r2', n_jobs=-1)
rfecv.fit(X,y)
```

```python
import matplotlib.pyplot as plt
plt.plot(rfecv.grid_scores_)
plt.title("Recursive feature elimination")
plt.xlabel("feature numbers")
plt.ylabel("R2")
```

```python
   #Ridge regression
    from sklearn.linear_model import LinearRegression, RidgeCV,Ridge
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
alphas=[0.001,0.01, 0.02, 0.05, 0.1, 0.2, 0.5, 1, 2, 5, 10, 20, 50, 100, 200, 500, 1000]
ridge_cofficients = []
for alpha in alphas:
    ridge = Ridge(alpha = alpha, normalize=True)
    ridge.fit(X, y)
    ridge_cofficients.append(ridge.coef_)
```

```python
plt.rcParams['axes.unicode_minus'] = False
plt.style.use('ggplot')
plt.plot(alphas, ridge_cofficients)
plt.xscale('log')
plt.axis('tight')
plt.title('Relationship between alpha values and Ridge cofficients')
plt.xlabel('Log Alpha')
plt.ylabel('Cofficients')
plt.show()
```

```python
ridge_cv = RidgeCV(alphas = alphas, normalize=True,cv = 5)
ridge_cv.fit(X, y)
```

```python
best_alpha = ridge_cv.alpha_
print(best_alpha)
ridge = Ridge(alpha = best_alpha, normalize=True)
ridge.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt
im = ridge.coef_
sorted_idx = np.argsort(im)

pos = np.arange(sorted_idx.shape[0])+0.5
im_sorted = im[sorted_idx]
name_sorted = np.array(X.columns)[sorted_idx]
plt.bar(pos, im_sorted )
plt.title("Ridge Regression")
plt.xlabel("features")
plt.ylabel("coef_")
```

```python
#lasso
from sklearn.linear_model import Lasso,LassoCV
alphas = 10**np.linspace(-6,-2,100)
lasso_cofficients = []

for alpha in alphas:
    lasso = Lasso(alpha = alpha,normalize=True)
    lasso.fit(X, y)
    lasso_cofficients.append(lasso.coef_)
    
plt.rcParams['axes.unicode_minus'] = False
# 设置绘图风格
plt.style.use('ggplot')
plt.plot(alphas, lasso_cofficients)
plt.xscale('log')
plt.axis('tight')
plt.title('Relationship between alpha values and Lasso cofficients')
plt.xlabel('Log Alpha')
plt.ylabel('Cofficients')
plt.show()
```

```python
lasso_cv = LassoCV(alphas = alphas, normalize=True, cv = 5)
lasso_cv.fit(X, y)
```

```python
best_alpha = lasso_cv.alpha_
print(best_alpha)
lasso = Lasso(alpha = best_alpha, normalize=True)
lasso.fit(X, y)
```

```python
from sklearn.linear_model import ElasticNetCV, ElasticNet
alphas = 10**np.linspace(-6,-1,100)
en_cofficients = []

for alpha in alphas:
    en = ElasticNet(alpha = alpha,normalize=True,l1_ratio=0.5)
    en.fit(X, y)
    en_cofficients.append(en.coef_)
    
plt.rcParams['axes.unicode_minus'] = False
# 设置绘图风格
plt.style.use('ggplot')
plt.plot(alphas, en_cofficients)
plt.xscale('log')
plt.axis('tight')
plt.title('Relationship between alpha values and ElasticNet cofficients')
plt.xlabel('Log Alpha')
plt.ylabel('Cofficients')
plt.show()
```

```python
alphas = 10**np.linspace(-6,-1,100)
en_cv =ElasticNetCV(alphas = alphas, normalize=True, cv = 5,l1_ratio=0.5)
en_cv.fit(X, y)
```

```python
selected = np.array(X.columns)[abs(en.coef_)>0.02]
```

```python
from sklearn.linear_model import LinearRegression
R2s=[]
coefs = []
adj_R2s=[]
for idx in selected:
    oneX = X[[idx]]
    lr = LinearRegression()
    lr.fit(oneX, y)
    R2 = lr.score(oneX, y)
    coef = lr.coef_[0]
    
    n=len(oneX)
    Adjust_R2=1-(1-R2)*(n-1)/(n-1-1)
    
    R2s.append(R2)
    coefs.append(coef)
    adj_R2s.append(Adjust_R2)
```

```python
#pd.set_option('precision', 2)
pd.set_option('display.float_format',lambda x : '%.3f' % x)
resultMT = pd.DataFrame({"features":list(selected), "R2":list(R2s), "adjR2":list(adj_R2s),"coefs":list(coefs)})
resultMT
```
