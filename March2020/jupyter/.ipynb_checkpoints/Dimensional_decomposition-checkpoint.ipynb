```python
import pandas as pd
import numpy as np
url="https://raw.githubusercontent.com/wangjk321/Intragenic-cohesin-project/master/March2020/bino161_normalizedContinuous15.matrix"
mt = pd.read_csv(url,sep="\t")
mt.index=mt.CohesinPos
mt.shape
mt.drop("CohesinPos",axis=1,inplace=True)
```

```python
#单独拿出用（-1，0，1）表示的Mvalue，不影响原来矩阵
Mvalue_tri = mt[["Mvalue"]]
Mvalue_tri_scaled = Mvalue_tri.apply(lambda x : (x-np.min(x))/(np.max(x)-np.min(x)))
Mvalue_tri_scaled.head()
#单独拿出来一个因变量，不影响原来矩阵
DI=(mt["Mvalue_continue"]<-0.5) & (mt["cohesin.intra"]==1)
```

```python
mt=mt.drop(["Mvalue"],axis=1)
mt.shape
```

```python
conti= mt.iloc[:,-15:]
bino= mt.iloc[:,:-15]
print(conti.shape, bino.shape)
```

```python
conti_scaled= conti.apply(lambda x : (x-np.min(x))/(np.max(x)-np.min(x)))
mt_scaled= pd.concat([bino,conti_scaled,Mvalue_tri_scaled],axis=1)
print(mt_scaled.shape)
```

```python
from sklearn import decomposition, datasets, model_selection, preprocessing, metrics
X=mt_scaled
```

```python
X
```

```python
#测试PCA的效果，比较原始矩阵X， 和矩阵W * H
from sklearn import metrics
def get_score(model, data, scorer=metrics.explained_variance_score):
    """ Estimate performance of the model on the data """
    prediction = model.inverse_transform(model.transform(data))
    return scorer(data, prediction)
```

```python
nc = 10
pca=decomposition.PCA(n_components=nc)
ica=decomposition.FastICA(n_components=nc)
nmf=decomposition.NMF(n_components=nc)
```

```python
pca.fit(X)
```

```python
ica.fit(X)
```

```python
nmf.fit(X)
```

```python
get_score(pca,X)
```

```python
get_score(ica,X)
```

```python
get_score(nmf,X)
```

```python
Wpca=pca.fit_transform(X)
Hpca=pca.components_

Wica=ica.fit_transform(X)
Hica=ica.components_

Wnmf=nmf.fit_transform(X)
Hnmf=nmf.components_
```

```python

def biplot(W,H,labels=None):
    score=W[:,0:2]
    coeff=np.transpose(H[0:2,:])
    xs = score[:,0] #PC1的值
    ys = score[:,1] #PC2的值
    n = coeff.shape[0] #feature的数量
    scalex= 1.0/(xs.max()-xs.min())
    scaley= 1.0/(ys.max()-xs.min())
    plt.scatter(xs*scalex,ys*scaley,c=DI,marker=".",s=0.11)
#    for i in range(n):
#        plt.arrow(0,0,coeff[i,0],coeff[i,1],color='r',alpha=0.5,head_width=0.05)
#        if labels is None:
 #           plt.text(coeff[i,0]*1.15, coeff[i,1]*1.15,"Var"+str(i+1),
 #                   color='g',ha='center',va='center')
 #       else:
 #           plt.text(coeff[i,0]*1.15, coeff[i,1]*1.15,labels[i],
  #                  color='g',ha='center',va='center')
    plt.xlim(-1,1)
    plt.ylim(-1,1)
    plt.xlabel("PC{}".format(1))
    plt.ylabel("PC{}".format(2))
    plt.grid()

biplot(Wpca,Hpca)
```

```python
def pca3Dplot(W,col):
    fig = plt.figure()
    ax = fig.gca(projection='3d')
    ax.scatter(W[:, 0], W[:, 1], W[:, 2], c=col,s=0.01)
    plt.show()
    
pca3Dplot(Wpca,DI)
```
